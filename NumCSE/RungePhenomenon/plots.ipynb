```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def runge(x):
    return 1/(1+25*x**2)
```

Coefficients for polynomial fit with linspace:

```python
# Coefficients for polynomial fit
a = [[1, -1.03309e-13, -16.8552, 2.12793e-12, 123.36, -1.11022e-11, -381.434, 1.87687e-11, 494.91, -9.69113e-12, -220.942]]

# Plot the polynomial
x = np.linspace(-1, 1, 100)
y = np.polyval(a[0][::-1], x)

plt.plot(x, y)
plt.plot(x, runge(x), label="Real function")
```
