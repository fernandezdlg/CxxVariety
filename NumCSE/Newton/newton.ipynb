```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# f0 == x**2 + 2*y**2 - 1
def f0(x,y):
    return x**2 + 2*y**2 - 1
```

```python
# f1 == y - x**2
def f1(x,y):
    return y - x**2
```

```python
x,y = np.meshgrid(np.linspace(-1,1,100),np.linspace(-1,1,100))
plt.contour(x,y,f0(x,y),[0], colors='r')
plt.contour(x,y,f1(x,y),[0], colors='b')
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
```
